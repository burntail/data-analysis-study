```python
#logisfic function (3-1 slide 34p)

def sigmoid(z):
    return 1/(1+np.exp(-z))

#h(x)

def hx(w,x):
    z=np.dot(x,w)
    return sigmoid(z)

#Loss function(3-1 slide p38)

def Loss_f(w,x,y):
    pred_y=hx(w,x)

    y=np.array(y)

    return np.sum(-y*np.log(pred_y)-(1-y)*np.log(1-pred_y))
   

```

```python

def gradient_decent(w,x,y,learning_rate,num_iter):
    
    y=np.array(y)
    
    for i in range(num_iter):
#         print(w)
        y_pred=hx(w,x)
        for j in range(3):
            w[j]=w[j]-learning_rate*np.sum((y_pred-y)*features[:,j])
    return w
    

```
